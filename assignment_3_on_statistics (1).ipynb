```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("/content/drive/MyDrive/IIT PALAKKAD ADV AI/data/StudentsPerformance.csv")
```

```python
#1.Calculate the mean, median, and mode for the math, reading, and writing scores.

print("Mean:\n", df[['math score','reading score','writing score']].mean())

print("\nMedian:\n", df[['math score','reading score','writing score']].median())

print("\nMode:\n", df[['math score','reading score','writing score']].mode().iloc[0])
```

```python
#2.Determine the range, variance, and standard deviation for each score category.
for col in ['math score','reading score','writing score']:
    print(f"\n--- {col} ---")
    print("Range:", df[col].max() - df[col].min())
    print("Variance:", df[col].var())
    print("Std Dev:", df[col].std())
```

```python
#3.Identify the mean, median, and mode of math scores for students with different parental education levels.
grouped = df.groupby('parental level of education')['math score']

print("\nMean:\n", grouped.mean())
print("\nMedian:\n", grouped.median())
print("\nMode:\n", grouped.apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan))
```

```python
#4.Compare the mean scores of students who took the test preparation course versus those who didn’t.

prep = df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
print("\nTest Prep Comparison:\n", prep)
```

```python
#5.Calculate the interquartile range (IQR) for each score category.lunch_std = df.groupby('lunch')['math score'].std()

for col in ['math score','reading score','writing score']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    print(f"{col} IQR:", IQR)
```

```python
#6.Determine the standard deviation for math scores by lunch type (free/reduced vs. standard).
lunch_std = df.groupby('lunch')['math score'].std()
print("\nStd Dev by Lunch:\n", lunch_std)
```

```python
#7.Find the correlation coefficient between math and reading scores, and then between reading and writing scores.
corr_math_read = df['math score'].corr(df['reading score'])
corr_read_write = df['reading score'].corr(df['writing score'])

print("\nCorrelation Math vs Reading:", corr_math_read)
print("Correlation Reading vs Writing:", corr_read_write)
```

```python
#8.Interpret the strength and direction of the correlation. What might this suggest about the relationship between different scores?
def interpret_corr(c):
    if c > 0.7:
        return "Strong Positive"
    elif c > 0.3:
        return "Moderate Positive"
    elif c > 0:
        return "Weak Positive"
    elif c < -0.7:
        return "Strong Negative"
    elif c < -0.3:
        return "Moderate Negative"
    else:
        return "Weak/No Relationship"

print("\nInterpretation:")
print("Math vs Reading:", interpret_corr(corr_math_read))
print("Reading vs Writing:", interpret_corr(corr_read_write))
```

```python
#9.Calculate the covariance between math and reading scores. How does it compare with the correlation? What does it suggest?
cov = df['math score'].cov(df['reading score'])
print("\nCovariance (Math vs Reading):", cov)
```

```python
#10.Z-Test: Assume the national average math score is 50. Use a one-sample z-test to test if this sample's average math score is significantly different from the national average.
sample_mean = df['math score'].mean()
pop_mean = 50
std = df['math score'].std()
n = len(df)

z = (sample_mean - pop_mean) / (std / np.sqrt(n))
print("\nZ-score:", z)

# Interpretation
if abs(z) > 1.96:
    print("Significant difference from population mean")
else:
    print("No significant difference")
```

```python
#11.T-Test: Compare the reading scores between students with different lunch types (free/reduced vs. standard) to test if there is a significant difference.
group1 = df[df['lunch'] == 'standard']['reading score']
group2 = df[df['lunch'] == 'free/reduced']['reading score']

t_stat, p_val = stats.ttest_ind(group1, group2)

print("\nT-Test:")
print("t-stat:", t_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Significant difference")
else:
    print("No significant difference")
```

```python
#12.F-Test: Check if the variances of math scores differ significantly between students with and without test preparation.
group1 = df[df['test preparation course'] == 'completed']['math score']
group2 = df[df['test preparation course'] == 'none']['math score']

f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
df1 = len(group1) - 1
df2 = len(group2) - 1

p_val = 1 - stats.f.cdf(f_stat, df1, df2)

print("\nF-Test:")
print("F-stat:", f_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Variances are significantly different")
else:
    print("Variances are not significantly different")
```
